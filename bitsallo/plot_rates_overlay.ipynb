```python
fig,(ax1,ax2) = plt.subplots(nrows=2,figsize=(12,12),gridspec_kw={'height_ratios':[5,1.5]})

ns2,bins,_ = ax1.hist(rates_2, bins=100, range=(0.,100), cumulative=-1,\
                      weights=weights_2, histtype='stepfilled', facecolor='blue', edgecolor='k', alpha=0.2,label=str(bits_2)+'bits')
ns7,bins,_ = ax1.hist(rates_1, bins=100, range=(0.,100), cumulative=-1,\
                      weights=weights_1, histtype='stepfilled', facecolor='green', edgecolor='k', alpha=0.2,label=str(bits_1)+'bits')
ax1.legend()

ax1.set_xlabel(r'$p_{T}$ threshold (GeV)', fontsize=15, horizontalalignment='right',x=1.0)
ax1.set_ylabel(r'Trigger rate (kHz)', fontsize=15, horizontalalignment='right',y=1.0)
ax1.set_xlim(0,60)
ax1.set_ylim(1e-1,1e4)
ax1.set_yscale('log')
ax1.tick_params(axis="y", labelsize=15)
ax1.tick_params(axis="x", labelsize=15)
ax1.text(28,10**3,'Work in Progress', fontsize=20,horizontalalignment='right',verticalalignment='top')
ax1.text(28,10**2.5,'1.65< $\eta$ <2.4', fontsize=20,horizontalalignment='right',verticalalignment='top')

ax2.plot(bins[:-1],ns2/ns7,alpha=0.4)
ax2.set_xlim(0,60)
ax2.set_ylim(0.5,1.5)
ax2.axhline(y=1.0, color='black', linestyle='dashed')

ax2.set_ylabel('%s bits / %s bits' %(bits_2, bits_1), fontsize=20,horizontalalignment='left',verticalalignment='bottom')
ax2.tick_params(axis="x", labelsize=15)


#plt.errorbar(xdata7, ydata7, xerr=xerr7, yerr=yerr7, color='#333333', ecolor='#333333', fmt='none', capsize=0, lw=1)
#plt.xlabel(r'$p_{T}$ threshold [GeV]')
#plt.ylabel(r'Trigger rate [kHz]')
#plt.xlim(0,30)
#plt.ylim(1e-1,1e4)
#plt.yscale('log')





plt.savefig('model_MK_'+str(bits_1)+'bits'+'_and_'+str(bits_2)+'bits_'+'rates_overlay'+'.pdf')
plt.show()

#logger.info("Rates in kHz at 18/19/20/21/22 GeV: {0}, {1}, {2}, {3}, {4}".format(hist[18], hist[19], hist[20], hist[21], hist[22]))
#print np.array2string(hist, separator=', ', precision=4, floatmode='fixed')
```
