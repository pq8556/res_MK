```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
plt.style.use('tdrstyle.mplstyle')
%matplotlib inline
```

```python
bits_1 = 5 # bigger bits
bits_2 = 34
gev = '15GeV'
```

```python
print('model_MK_'+str(bits_1)+'bits_eff_'+'xdata'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'xerr'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'ydata'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'yerr'+gev+ '.npy')
print('#####################################################')
print('model_MK_'+str(bits_1)+'bits_eff_'+'xdata'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'xerr'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'ydata'+gev+ '.npy')
print('model_MK_'+str(bits_1)+'bits_eff_'+'yerr'+gev+ '.npy')

xdata_1 = np.load('model_MK_'+str(bits_1)+'bits_eff_'+'xdata'+gev+ '.npy') # xdata
xerr_1 = np.load('model_MK_'+str(bits_1)+'bits_eff_'+'xerr'+gev+ '.npy') # xerr
ydata_1 = np.load('model_MK_'+str(bits_1)+'bits_eff_'+'ydata'+gev+ '.npy') # ydata
yerr_1 = np.load('model_MK_'+str(bits_1)+'bits_eff_'+'yerr'+gev+ '.npy') # yerr

xdata_2 = np.load('model_MK_'+str(bits_2)+'bits_eff_'+'xdata'+gev+ '.npy') # xdata
xerr_2 = np.load('model_MK_'+str(bits_2)+'bits_eff_'+'xerr'+gev+ '.npy') # xerr
ydata_2 = np.load('model_MK_'+str(bits_2)+'bits_eff_'+'ydata'+gev+ '.npy') # ydata
yerr_2 = np.load('model_MK_'+str(bits_2)+'bits_eff_'+'yerr'+gev+ '.npy') # yerr
```

```python
plt.figure(figsize=(12,9))
# , marker=',' capsize=0, , lw=1
lines = {'linestyle': 'None'}
plt.rc('lines', **lines)
plt.errorbar(xdata_1, ydata_1, xerr=xerr_1, yerr=yerr_1, color='green', label=str(bits_1)+'bits' , marker='o')
plt.errorbar(xdata_2, ydata_2, xerr=xerr_2, yerr=yerr_2, color='blue', label=str(bits_2)+'bits', marker='o')
plt.axhline(y=1.0, color='r', linestyle='-')
dictCut = {}
dictCut['15GeV'] = 15.0
dictCut['25GeV'] = 25.0
plt.axvline(x=dictCut[gev], ymin=0, ymax=0.91, color='black', linestyle='dashed')
plt.plot([0,120], [0.9,0.9], ls='dashed', lw=0.5, color='black')
plt.text(110,0.8,'Work in Progress', fontsize=20,horizontalalignment='right',verticalalignment='top')
plt.text(110,0.7,'1.65< $\eta$ <2.4', fontsize=20,horizontalalignment='right',verticalalignment='top')
plt.xlim(0,120)
plt.ylim(0,1.1)
plt.tick_params(axis="y", labelsize=15)
plt.tick_params(axis="x", labelsize=15)
plt.xlabel(r'$p_{T}$ (GeV)', fontsize=15, horizontalalignment='right',x=1.0)
plt.ylabel('Efficiency', fontsize=15, horizontalalignment='right',y=1.0)
plt.legend()
plt.savefig('model_MK_'+str(bits_1)+'bits'+'_and_'+str(bits_2)+'bits'+'_eff_'+'xdata'+gev+ '.pdf')
plt.show()
```
